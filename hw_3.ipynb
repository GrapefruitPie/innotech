Исправим опечатки и сокращения в названиях машин.

```python
df.CarName = df.CarName.str.lower()
```

```python
df['mark'] = df.CarName.map(lambda x: x.split()[0])
pd.Series(df.mark).unique()
```

```python
df.loc[(df.mark == 'maxda'), 'mark'] = 'mazda'
df.loc[(df.mark == 'porcshce'), 'mark'] = 'porsche'
df.loc[(df.mark == 'toyouta'), 'mark'] = 'toyota'
df.loc[(df.mark == 'vokswagen')|(df.mark == 'vw'), 'mark'] = 'volkswagen'
pd.Series(df.mark).unique()
```

Для простоты уберем из названия модели тире и склеим все символы.

```python
df['model'] = df.CarName.map(lambda x: str(''.join(x.split()[1:])).replace('-', ''))
pd.Series(df.model).unique()
```

```python
df.CarName = df[['mark', 'model']].apply(
    lambda x: ' '.join(x.dropna().astype(str)),
    axis=1
)
df.CarName.head()
```

```python
df = df.drop(columns=['mark', 'model'])
df.drop_duplicates(inplace=True)
df.shape
```

Заменим категориальные значения на численные.

```python
cylinders = {'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'eight': 8, 'twelve': 12}
for key, value in cylinders.items():
    df.loc[(df.cylindernumber == key), 'cylindernumber'] = value
df.cylindernumber = df.cylindernumber.astype('int64')
```

```python
data = df.drop(columns=['car_ID', 'carlength', 'carwidth', 'carheight', 'curbweight', 'citympg', 'doornumber', 'enginelocation']).copy()
```

Переведем все категориальные признаки в бинарные.

```python
datatypes = data.columns.to_series().groupby(df.dtypes).groups
datatypes
```

```python
ohe = OneHotEncoder()
for column in ['CarName', 'fueltype', 'aspiration', 'carbody', 'drivewheel', 'enginetype', 'fuelsystem']:
    column_ohe = ohe.fit_transform(data[[column]]).toarray()
    column_ohe_df = pd.DataFrame(column_ohe, columns = ohe.categories_[0])
    column_ohe_df.index = data.index
    data = data.join(column_ohe_df)
```

```python
data = data.drop(columns=['CarName', 'fueltype', 'aspiration', 'carbody', 'drivewheel', 'enginetype', 'fuelsystem'])
data.head()
```

Обучим модель линейной регрессии.

```python
model = linear_model.SGDRegressor(alpha = 0.01, max_iter = 1000)
X = data.drop(columns='price')
y = data.price

model.fit(X, y)
```

```python
model.intercept_, model.coef_
```

```python
estimator = linear_model.LinearRegression()
```

```python
estimator.fit(X, y)
estimator.intercept_, estimator.coef_
```

```python
y_predicted = estimator.predict(X)

print(f'MAE: {mean_absolute_error(y, y_predicted)}')
print(f'MSE: {mean_squared_error(y, y_predicted)}')
print(f'RMSE: {mean_squared_error(y, y_predicted) ** 0.5}')
```
